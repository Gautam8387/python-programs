```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
def plot(x_str, _run, num_features, time_taken_wall, time_taken_usr, time_taken_sys, title):
    # Plot
    x_str
    sns.set_theme(style='whitegrid')
    plt.figure(figsize=(7, 5))
    sns.lineplot(x=x_str, y='Wall Time (s)', data=_run, color='red')
    plt.title(f'Run Time: {title}')
    plt.xlabel(f'{x_str} (log10)')
    plt.ylabel('Time Taken (seconds)')
    plt.xscale('log')
    # Plot red dots at the data points
    plt.scatter(num_features, time_taken_wall, color='black')

    # Dotted plots for User and System Time
    plt.plot(num_features, time_taken_usr, linestyle='dotted', color='black')
    plt.plot(num_features, time_taken_sys, linestyle='dotted', color='green')

    legend_data = {'Wall Time': 'red', 'User Time': 'black', 'System Time': 'green'}
    legend_labels = list(legend_data.keys())
    legend_handles = [plt.Line2D([0], [0], color=color, marker='o', linestyle='') for color in legend_data.values()]
    plt.legend(legend_handles, legend_labels, title='Time Type', loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
# Feature Run
num_features = [10, 100, 1000, 10000, 100000]
time_taken_wall = [0.061, 0.203, 0.502, 0.840, 5.2]
time_taken_usr = [0.043, 0.375, 1.2, 1.48, 5.82]
time_taken_sys = [0.012, 0.94, 0.171, 0.252, 0.170]

feature_run = pd.DataFrame({'Number of Features': num_features, 'Wall Time (s)': time_taken_wall, 'User Time (s)': time_taken_usr, 'System Time (s)': time_taken_sys })
feature_run
```

```python
plot('Number of Features', feature_run, num_features, time_taken_wall, time_taken_usr, time_taken_sys, 'Feature (V1)')
```

```python
def plot_2(feature_run, feature_run_old, feature_run_new, num_features):

    plt.figure(figsize=(7, 5))
    sns.lineplot(x='Number of Features', y='V1 Wall Time (s)', data=feature_run, color='red')
    sns.lineplot(x='Number of Features', y='V2 Wall Time (s)', data=feature_run, color='blue')
    plt.title('Feature Run Time')
    plt.xlabel('Number of Features (log10)')
    plt.ylabel('Time Taken (seconds)')
    plt.xscale('log')

    # Plot red dots at the data points
    plt.scatter(num_features, feature_run_new, color='black')
    plt.scatter(num_features, feature_run_old, color='black')

    legend_data = {'V1 Wall Time': 'red', 'V2 Wall Time': 'blue'}
    legend_labels = list(legend_data.keys())
    legend_handles = [plt.Line2D([0], [0], color=color, marker='o', linestyle='') for color in legend_data.values()]
    plt.legend(legend_handles, legend_labels, title='Time Type', loc='upper left')

    plt.tight_layout()
    plt.show()
```
